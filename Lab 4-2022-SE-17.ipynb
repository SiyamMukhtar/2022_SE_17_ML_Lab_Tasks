```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.histplot(df["Aggregate rating"], bins=20, kde=True, color="blue")
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_cuisines = df["Cuisines"].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="coolwarm")
plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Price range", data=df, palette="viridis")
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="Price range", y="Average Cost for two", data=df, palette="muted")
plt.ylim(0, 5000)  #Limit
plt.title("Average Cost for Two Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Average Cost for Two")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Votes", y="Aggregate rating", data=df, alpha=0.5)
plt.title("Votes vs. Aggregate Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
#Key Insights from the Analysis
#Ratings Distribution: Most restaurants have ratings between 3.0 and 4.5.
#Popular Cuisines: The top 10 most common cuisines indicate strong market preferences.
#Price Range Distribution: Mid-range restaurants are the most common.
#Cost Variation: Higher price ranges have broader cost variation.
#Votes and Ratings: More votes generally correlate with higher ratings.
#Dining Cost: The cost for two varies even within the same price range.
#Cuisine Popularity: Certain cuisines are significantly more common.
#Online Delivery: Many restaurants do not offer online delivery, indicating a dine-in preference.
#Customer Engagement: Higher-rated restaurants tend to receive more votes.
#Cuisines and Ratings: Some cuisines may be associated with higher ratings.
```
